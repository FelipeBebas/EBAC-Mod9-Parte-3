### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
#separar colunas do exercicio
tarefa_2 = sinasc_raw[['DTNASC','PESO','SEXO','ESCMAE']]
```

```python
#transformar coluna DTNASC para datetime
sinasc_raw['DTNASC'] = pd.to_datetime(tarefa_2['DTNASC'])
```

```python
#filtrar data de aniversário
aniversario = tarefa_2[(sinasc_raw['DTNASC'].dt.month == 9) & (sinasc_raw['DTNASC'].dt.day == 25)]
```

```python
tarefa_2_resultado = aniversario.groupby(['ESCMAE', 'SEXO'])['PESO'].mean().reset_index()
```
